```python
# 爬房仲網公用爬蟲模組
import requests as r
from bs4 import BeautifulSoup as bs
import lxml
from datetime import datetime
import sqlite3
import json
import sys
import re
import math
import random


# 從db抓取 已存入(但尚未被request內文)的網址清單
def qryNullUrls(dbName):
    with sqlite3.connect(dbName) as conn:
        c = conn.cursor()
        qry = "SELECT url from rental WHERE title is null"
        c.execute(qry)
    return c.fetchall()
    conn.close()
#


# 取得地區資訊json -> list
# 讀取資料夾底下的json檔, 裡面記載著 
# {"都/縣/市":["鄉/鎮/市/區", "鄉/鎮/市/區", ...] }
# 的資訊, 會回傳 ex: 
# [[台北市, 信義區],
# [台北市, 北投區],
# [台北市, 文山區]]
# 的list
def getLocData(importFile):
    with open(importFile, 'r', encoding='utf8') as f:
        loc = json.loads(f.read())
    f.close()
    lst = []
    
    # 將地區資訊, 轉換成list資訊
    for city in loc.keys():
        for area in loc[city]:
            lst.append([city, area])
    return lst
#


# 0. 抓特定外頁的總筆數 -> 總頁數
def getTotalPage(dbName, url, head="", payload=""):
    try:
        # 信義房屋
        if dbName == 'sy.db':
            res = r.get(url)
            soup = bs(res.text, 'lxml')
            nums = soup.select('h1')[0].text
            regex = re.compile('\d*[,]?\d+')
            nums = int(regex.search(nums).group().replace(',', ''))
            totalPage = math.ceil(int(nums) / 20)
            return totalPage
        
        # 永慶房屋
        elif dbName == 'yc.db':
            res = r.post(url, headers = head, data = payload)
            soup = bs(res.text, 'lxml')
            return int(soup.select('#hidTotalPages')[0]['value'])
        
        # 中信房屋
        elif dbName == 'jc.db':
            res = r.post(url, headers = head, data = payload)
            soup = bs(res.text, 'lxml')
            a = soup.select('p')[0].text
            p = re.compile('"totalCount":[\d]+')
            qq = p.findall(a)[0]
            tick = int(qq.split(':')[1])
            return math.ceil(tick / 10)
            
        elif dbName == 'hf.db':
            pass
            
        else:
            print('沒有這個db...' + dbName + 'public_crawler.getTotalPage')
            sys.exit()
    except Exception as err:
        print('public_crawler.getTotalPage Error in ' + dbName + ' with ' + url)
        sys.exit()
# 

        
# 1.外頁網址, 依照不同房仲網, 解析出各自的urls
def extractAllUrls(dbName, soup, index=""):
    try:
        allUrls = []
        
        #  信義房屋 # 
        if dbName == 'sy.db':
            for url_raw in soup.select('a.name'):
                url = url_raw['href']
                allUrls.append(url)
            return allUrls
            

        # 永慶房屋、好房網 #
        elif dbName == 'yc.db':
            urls_yc = []
            urls_hf = []
            try:
                tag_href = soup.select('div.house_block_content h2 a')
                for tag_href in tag_href:
                        # 若tag中有包含 'http' 字樣, 表示此網址為 "好房網" , 否則為 "永慶房屋"
                    if 'http' not in tag_href['href']:
                        url = index + tag_href['href']
                        urls_yc.append(url)

                    else:
                        url = tag_href['href']
                        urls_hf.append(url)

            except Exception as err:
                print(str(err) + ' ; 網頁頁面網改?   or 程式BUG?')
                sys.exit()
            
            return urls_yc, urls_hf
        

        # 中信房屋 #
        elif dbName == 'jc.db':
            idRegex = re.compile(r'id=\\"[\d]+\\"')
            tags = idRegex.findall(soup.text)
            for tag in tags:
                idRegex = re.compile(r'[\d]+')
                iid = idRegex.findall(tag)[0]
                href = index + '/Building/Detail/' + iid
                allUrls.append(href)
            return allUrls
            
        else:
            print('沒有這個db...' + dbName + 'public_crawler.extractAllUrls')
 
    except Exception as err:
        print('解析allUrls Error' + dbName)
#


# 2.特定頁面所有網址, 到資料庫看有沒有儲存過, 沒有的話再把他堆疊, 最後回傳
def filterUrls(dbName, allUrls):
    try:
        urls = []
        
        for url in allUrls:
            if saveThisUrl(dbName, url):
                urls.append(url)

            else:
                updateGetTime(dbName, url)

        return urls

        
    except Exception as err:
        print('解析urls Error' + dbName)
        sys.exit()
#


# 3.外網解析時, 抓到網址清單, 逐筆檢查此url是否已有
def saveThisUrl(dbName, url):
    try:
        with sqlite3.connect(dbName) as conn:
            c = conn.cursor()
            c.execute("SELECT url FROM rental WHERE url = '" + url + "'")
            if (len(c.fetchall()) == 0):
                return True
            else:
                return False
    except Exception as err:
        print('Repeat check error with(不明錯誤) : ' + url)
#


# 4.如果已有此份網址, 將 getTime更新為現在時間
def updateGetTime(dbName, url):
    try:
        with sqlite3.connect(dbName) as conn:
            c = conn.cursor()
            getTime = str(datetime.now())
            c.execute("UPDATE rental SET getTime = '" + getTime + "' WHERE url = '" + url + "'")
    except Exception as err:
        print('Something error with update getTime with ' + url)
        print(str(err))
        sys.exit()
    finally:
        conn.close()
#

   
# 儲存 url, title, main, detail, getTime -> sqlite
def insertToDB(dbName, url, title = '', main = '', detail = ''):
    try:
        with sqlite3.connect(dbName) as conn:

            c = conn.cursor()
            getTime = str(datetime.now())

            if title == '':
                insertStr = 'INSERT INTO rental(url, getTime) VALUES (?, ?)'
                values = (url, getTime)
                
            else:
                insertStr = 'INSERT INTO rental(url, title, main, detail, getTime) VALUES (?, ?, ?, ?, ?)'
                values = (url, title, main, detail, getTime)

            c.execute(insertStr, values)
            num = 1

    except Exception as err:
        print('url repeat in ' + dbName + ' with ' + url)
        num =  0
    finally:
        conn.close()
        print(num)
        return num
#


def updateToDB(dbName, url, title, main, detail):
    try:
        with sqlite3.connect(dbName) as conn:
            c = conn.cursor()
            # 逐筆查詢資料庫是否存過此網址

            getTime = str(datetime.now())
            
            qryString = 'UPDATE rental SET title=?, main=?, detail=?, getTime=? WHERE url=?' 
            c.execute(qryString, (title, main, detail, getTime, url))

        conn.close()
    except Exception as err:
        print('updateToDB Error ' + url)
#


# 從檔案中選擇已存的ip清單, 隨機取一筆      
def getRandomIP():
    with open('ip.txt', 'r') as f:
        ips = f.read()
    f.close()
    ips = ips.split(',')
    return ips[random.randint(0, len(ips) - 1)]
#
```

```python
## 將proxy轉成python格式
def ipList2pyFormat(ipList):
    result = []
    if isinstance(ipList, list):
        for ip in ipList:
            result.append(ip2pyFormat(ip))
        return result
    else:
        return ip2pyFormat(ipList)
        

def ip2pyFormat(ip):
    ipDict = {}
    ipDict['http'] = 'http://' + ip
    return str(ipDict)


## 取得 5組proxy
#!pip install PySocks
def getIPfromPySocks():
    proxyLoop = []
    proxyHref = 'https://premproxy.com/proxy-by-country/Taiwan-01.htm'
    proxyRes = requests.get(proxyHref)
    proxySoup = bs(proxyRes.text,'lxml')
    proxySel = proxySoup.select('#proxylist > tr > td')
    for i in range(0,len(proxySel)):
        if ( i % 6 == 0):
            ip = proxySel[i].text
            proxyLoop.append(ip)
    return proxyLoop
```
