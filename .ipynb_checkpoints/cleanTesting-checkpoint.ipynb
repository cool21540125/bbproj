```python
# 永慶房屋, 信義房屋, 中信房屋
def readDB(target):
    with db.connect(target) as conn:
        c = conn.cursor()
        c.execute('select url, title, main, detail, getTime from rental;')
        result = c.fetchall()
    conn.close()
    return result
#
```

```python
# 住商不動產
def readDB_lb(target):
    with db.connect(target) as conn:
        c = conn.cursor()
        c.execute("select url, title, address, pattern, ttype, avg_rent, lat, lng, sex, space, smoke, pet, cook, parking, updateDate FROM rental;")
        result = c.fetchall()
    conn.close()
    return result
#
```

```python
# 永慶房屋, 信義房屋, 中信房屋    db欄位的拆解方法
def splitColumnFromDataSet(tbl):
    urls = []; titles = []; mains = []; details = []; getTimes = []; 
    for r in tbl:
        if r[0] != None:
            urls.append(r[0])
            titles.append(r[1])
            mains.append(r[2])
            details.append(r[3])
            getTimes.append(r[4])
    return urls, titles, mains, details, getTimes
#
```

```python
# 住商不動產  db欄位拆解方法
def splitColumnFromDataSet_lb(tbl):
    urls = [];        titles = [];     addresss = [];     patterns = [];     ttypes = []; 
    avg_rents = [];   lats = [];       lngs = [];         sexs = [];         spaces = [];
    smokes = [];      pets = [];       cooks = [];        parkings = [];     updateDates = []
    for r in tbl:
        if r[0] != None:
            urls.append(r[0])
            titles.append(r[1])
            addresss.append(r[2])
            patterns.append(r[3])
            ttypes.append(r[4])
            avg_rents.append(r[5])
            lats.append(r[6])
            lngs.append(r[7])
            sexs.append(r[8])
            spaces.append(r[9])
            smokes.append(r[10])
            pets.append(r[11])
            cooks.append(r[12])
            parkings.append(r[13])
            updateDates.append(r[14])
    return urls, titles, addresss, patterns, ttypes, avg_rents, lats, lngs, sexs, spaces, smokes, pets, cooks, parkings, updateDates
#
```

```python
# 讀取外部組態檔
def getConfig(cfgini):
    config = configparser.ConfigParser()
    config.read(cfgini)
    dbhost = config.get('Section_A', 'sql_dbhost')
    user = config.get('Section_A', 'sql_id')
    passwd = config.get('Section_A', 'sql_pd')
    return dbhost, user, passwd
#
```

```python
# 更新修改 mySQL資料庫
def upsertHouse(h, dbhost, user, passwd):
    try:
        db = pymysql.connect(dbhost, user, passwd, 'house')
        c = db.cursor()
        c.execute("SELECT url FROM rental WHERE url = %s;", (h.getUrl,))
        if (len(c.fetchall()) == 0):
            job = 'insert'
            
            doing = "INSERT INTO rental(url, floor, stories, ttype, avg_rent, lat, lng, sex, space, smoke, pet, cook, parking, updateDate) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
            vv = (h.getUrl, str(h.getFloor), str(h.getStories), h.getLabel, str(h.getAvg_rent), h.getLat, h.getLng, h.getSex, str(h.getSpace), h.getSmoke, h.getPet, h.getCook, h.getParking, h.getDate)
            print(vv)
            print('leave upsert')
            sys.exit()
            c.execute(doing, vv)
            db.commit()
            
        else:
            job = 'update'
            doing = "UPDATE rental SET updateDate = %s  WHERE url = %s ;"
            vv = (h.getDate, h.getUrl)

            c.execute(doing, vv)
            db.commit()

    except Exception as err:
        print('error: upsertHouse() with ' + h.getUrl + ' on ' + str(h.getDate) + ' when ' + job) 
        print(str(err))

    finally:
        c.close()
        db.close()
#
```

```python
# 信義房屋_建彰
class Extracter_sy():
    pass

    def __init__(self, url, title, main, detail, getTime):
        self.__url = url
        self.__title = title
        self.__main = main
        self.__detail = detail
        self.__getTime = getTime
        self.__YN = True
        
    @property # 調整的儲存程序
    def getYN(self):
        return self.__YN
        
    @property
    def getUrl(self):                            # OK
        return self.__url
    
    @property
    def getTitle(self):                          # OK
        return self.__title
    
    @property
    def getDate(self):
        d = self.__getTime.split(' ')[0]
        fmt = '%Y-%m-%d'
        return time.strptime(d, fmt)
    
    @property
    def getLabel(self):
        reg = re.compile(r'^[1-9][0-9]*房')      #  需要大量debug
        
        if self.__detail['類型'] == '店面':
            return 'S'
        elif self.__detail['類型'] == '住辦':
            return 'O'
        elif self.__detail['類型'] in ['透天厝/別墅', '雅房', '套房', '整層住宅']:
            return 'H'
        elif self.__detail['類型'] in ['辦公/廠房', '車位', '其它', '土地/場地']:
            return 'F'
        else:
            return 'X'
    
    @property
    def getFloor(self):                          # 所在樓層 @@
        try:
            if self.__main.get('floor'):
                ff = self.__main['floor'].split('/')[0]
                reg = re.compile(r'[\d]+')
                ff1 = reg.search(ff)
                if ff1.group() == 0:
                    return 1
                print(ff1.group())
                return ff1.group()
            else:
                return 0
        except Exception as err:
            return 0
    
    @property
    def getStories(self):                       # 總樓層 @@
        try:
            if self.__main.get('floor'):
                ft = self.__main['floor'].split('/')[1]
                reg = re.compile(r'[\d]+')
                ff2 = reg.search(ft)
                return ff2.group()
            else:
                return 0
        except Exception as err:
            return 0
    
    @property
    def getPattern(self):                        # OK
        if self.__detail['格局']:
            return self.__detail['格局']
        return ''
    
    @property
    def getAddress(self):                        # OK
        if self.__detail['地址']:
            return self.__detail['地址'].replace('　','')
        return ''
    
    @property
    def getAvg_rent(self):                       # 需要大量debug
        reg = re.compile(r'([0-9]{1,3},?[0-9]{3})+')
        if reg.search(self.__main['rent']):
            mo = reg.match(self.__main['rent'])
            return int(mo.group().replace(',', ''))
        return -1
    
    @property
    def getLat(self):                            # OK
        return self.__main['latlng'].split(',')[1]
        
    
    @property
    def getLng(self):                            # OK
        return self.__main['latlng'].split(',')[0]
    
    @property
    def getSex(self):                            # 需要大量debug
        if self.__detail.get('身份限制'):
            if '不拘' in self.__detail['身份限制']:
                return 'B'
            elif '限男性' in self.__detail['身份限制']:
                return 'M'
            elif '限女性' in self.__detail['身份限制']:
                return 'F'
        return ''

    @property
    def getSpace(self):                          # 需要大量debug
        if self.__detail.get('建物坪數'):
            mo = re.match(r'\d+(.[\d]+)?', self.__detail['建物坪數'])
            sp = float(mo.group())
            if sp > 250:         # 坪數太大者, 直接忽略
                self.__YN = False
            return sp
        else:
            return -1
    
    @property
    def getSmoke(self):                          # 需要大量debug
        if self.__detail.get('生活約定'):
            if '抽菸不可' in self.__detail['生活約定']:
                return 'N'
            elif '抽菸可' in self.__detail['生活約定']:
                return 'Y'
        else:
            return ''
    
    @property
    def getPet(self):                            # 需要大量debug
        if self.__detail.get('生活約定'):
            if '養寵物不可' in self.__detail['生活約定']:
                return 'N'
            elif '養寵物可' in self.__detail['生活約定']:
                return 'Y'
        else:
            return ''
    
    @property
    def getCook(self):                           # 需要大量debug
        if self.__detail.get('生活約定'):
            if '開伙不可' in self.__detail['生活約定']:
                return 'N'
            elif '開伙可' in self.__detail['生活約定']:
                return 'Y'
        else:
            return ''
    
    @property
    def getParking(self):                        # 需要大量debug
        
        if self.__detail.get('車位'):
            if '年' in self.__detail['車位']:
                self.__YN = False
                return 'X'
            if '無' == self.__detail['車位']:
                return 'N'
            elif self.__detail['車位'] in ['其他', '另租', '坡道平面', '坡道機械', '庭院', '有', '機械平面']:
                return 'Y'
        else:
            return ''
#
```

```python
# 住商不動產_建彰+逸帆
class Extracter_lb():
    pass

    def __init__(self, url, title, address, pattern, ttype, avg_rent, lat, lng, sex, space, smoke, pet, cook, parking, updateDate):
        self.__url = url
        self.__title = title
        self.__address = address
        self.__pattern = pattern
        self.__ttype = ttype
        self.__avg_rent = avg_rent
        self.__lat = lat
        self.__lng = lng
        self.__sex = sex
        self.__space = space
        self.__smoke = smoke
        self.__pet = pet
        self.__cook = cook
        self.__parking = parking
        self.__updateDate = updateDate
        
    @property
    def getUrl(self):                            # OK
        return self.__url
    
    @property
    def getTitle(self):                          # OK
        return self.__title
    
    @property
    def getAddress(self):                        # OK
        return self.__address
    
    @property
    def getPattern(self):                        # OK
        return self.__pattern
    
    @property
    def getLabel(self):
        return self.__ttype

    @property
    def getDate(self):
        return self.__updateDate

    @property
    def getAvg_rent(self):
        return self.__avg_rent
    
    @property
    def getLat(self):                            # OK
        return self.__lat

    @property
    def getLng(self):                            # OK
        return self.__lng
    
    @property
    def getSex(self):                            # 
        if self.__sex == None:
            return ''
        return self.__sex
    
    @property
    def getSpace(self):                          #
        if self.__space == None:
            return -1
        return self.__space
    
    @property
    def getSmoke(self):                          # 
        if self.__smoke == None:
            return ''
        return self.__smoke
    
    @property
    def getPet(self):                            # 
        if self.__pet == None:
            return ''
        return self.__pet
    
    @property
    def getCook(self):                           # 
        if self.__cook == None:
            return ''
        return self.__cook
    
    @property
    def getParking(self):                        # 
        if self.__parking == None:
            return ''
        return self.__parking
#
```

```python
# 寫出至json, 準備匯入mongoDB
def append2mongoJSON(fileName, obj):
    obj = str(obj)
    with open(fileName, 'a', encoding='utf8') as f:
        f.write(obj + "\n")
#
```

```python
# 產生可逐筆 append 到 mongoFormatJSON 的物件
def writeMongoFormat(houseClass):
    obj = {}
    obj['url'] = houseClass.getUrl
    obj['title'] = houseClass.getTitle
    obj['address'] = houseClass.getAddress
    obj['pattern'] = houseClass.getPattern
    return obj
#
```

```python
# 資料異常調整
def adj(dbhost, user, passwd):
    db = pymysql.connect(dbhost, user, passwd, 'house')
    c = db.cursor()
    
    # to do list:
    updQrys = []
    updQrys.append("UPDATE rental SET lat = -1, lng = -1 WHERE lat > 9999 OR lng > 9999;")      # 1. 更改 lat, lng過大者->-1
    updQrys.append("UPDATE rental SET lat = -1, lng = -1 WHERE lat <= 0 OR lng <= 0;")          # 2. 更改 lat = 0 -> -1
    updQrys.append("UPDATE rental SET space = -1 WHERE space = 0;")                             # 3. 更改 space = 0 -> -1
    updQrys.append("DELETE FROM rental WHERE avg_rent < 500;")                                  # 4. 刪除 avg_rent < 500
    updQrys.append("DELETE FROM rental WHERE parking = 'X';")                                   # 5. 刪除 parking不合理者

    for qry in updQrys:
        c.execute(qry)
        db.commit()
        
    c.close()
    db.close()
#
```

```python
def mainProg(target):
#     os.chdir('../')
    os.chdir(r'D:/bbproj/house')
    cfgini = 'params/house.ini'
    dbhost, user, passwd = getConfig(cfgini)
    fileName = 'mongo_' + target[0:2] + '.txt'
    os.remove(fileName) if os.path.exists(fileName) else None
    
    if target in ['sy.db', 'jc.db', 'yc.db']:
        tbl = readDB(target)
        urls, titles, mains, details, getTimes = splitColumnFromDataSet(tbl)
        
        for i in range(0, len(urls)):
            url = urls[i]
            title = titles[i]
            main = eval(mains[i])
            detail = eval(details[i])
            getTime = getTimes[i]

            if target == 'sy.db':
                houseClass = Extracter_sy(url, title, main, detail, getTime)  # 信義房屋
            elif target == 'jc.db':
                houseClass = Extracter_jc(url, title, main, detail, getTime) # 永慶房屋
            elif target == 'yc.db':
                houseClass = Extracter_yc(url, title, main, detail, getTime) # 永慶房屋
            
            upsertHouse(houseClass, dbhost, user, passwd)
            obj = writeMongoFormat(houseClass)
            qq = json.dumps(obj, ensure_ascii=False)
            append2mongoJSON(fileName, qq)

            
    elif target == 'lb.db':
        tbl = readDB_lb(target)
        urls, titles, addresss, patterns, ttypes, avg_rents, lats, lngs, sexs, spaces, smokes, pets, cooks, parkings, updateDates = splitColumnFromDataSet_lb(tbl)

        for i in range(0, len(urls)):
            url = urls[i];            title = titles[i];  address = addresss[i];  pattern = patterns[i];  ttype = ttypes[i]
            avg_rent = avg_rents[i];  lat = lats[i];      lng = lngs[i];          sex = sexs[i];          space = spaces[i]
            smoke = smokes[i];        pet = pets[i];      cook = cooks[i];        parking = parkings[i];  updateDate = updateDates[i]

            houseClass = Extracter_lb(url, title, address, pattern, ttype, avg_rent, lat, lng, sex, space, smoke, pet, cook, parking, updateDate)

            upsertHouse(houseClass, dbhost, user, passwd)
            obj = writeMongoFormat(houseClass)
            qq = json.dumps(obj, ensure_ascii=False)
            append2mongoJSON(fileName, qq)
            
    adj(dbhost, user, passwd)
#
```

```python
def gogo():
#     targets = ['sy.db', 'jc.db', 'yc.db', 'lb.db']
#     targets = ['lb.db']
    targets = ['sy.db']
    for target in targets:
        mainProg(target)
#
```
