```python
# 地址轉換經緯度
def tran(address):
    google_places = GooglePlaces(your_key)
    lat_lng = google_places.text_search(address).places[0].geo_location
    lat = float(lat_lng['lat'])
    lng = float(lat_lng['lng'])
    return lat, lng
```

```python
addressRegex = re.compile('.+[市縣](.+[鄉鎮市區]?)?(.*段?)?.+[路街巷弄].+[街弄號]')
```

```python
def updLatLng(url, lat, lng):
    db = pymysql.connect(host = 'localhost', port = 3306, user = 'crawler', passwd = '12345687', db = 'house', charset='utf8mb4')
    c = db.cursor()
    vv = (lat, lng, url)
    c.execute("UPDATE rental SET lat = %s, lng = %s WHERE url = %s", vv)
    db.commit()
    c.close()
    db.close()
```
